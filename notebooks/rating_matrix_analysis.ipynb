```python
import pandas as pd

# df = pd.read_csv('./csv_output/ratingInfo.csv')
df = pd.read_csv('./csv_output/commentsInfo_flair.csv')
df = df[~df['comments'].str.contains('This is an automated posting.')]
df
```

```python
duplicates = df[df.duplicated(subset=['userID', 'listing_id'], keep=False)]
duplicates
```

```python
df = df.groupby(['userID', 'listing_id']).agg({'estimated_rating_flair': 'mean'}).reset_index()
df
```

```python
# rating_matrix = df.pivot(index='userID', columns='listing_id', values='rating')
rating_matrix = df.pivot(index='userID', columns='listing_id', values='estimated_rating_flair')
rating_matrix.head()
```

```python
rating_matrix = rating_matrix.fillna(0)
rating_matrix
```

```python
def recommend_method(sparsity):
    if sparsity > 0.99:
        return "The matrix is very sparse. It is recommended to use matrix factorization methods such as SVD or ALS."
    elif sparsity > 0.95:
        return "The matrix is moderately sparse. Matrix factorization methods might be more suitable, but collaborative filtering methods could also perform well with proper parameter tuning."
    else:
        return "The matrix is not very sparse. Collaborative filtering methods such as user-user or item-item collaborative filtering could perform well."

# Calculate sparsity
sparsity = 1.0 - (rating_matrix > 0).sum().sum() / (rating_matrix.shape[0] * rating_matrix.shape[1])
print('Matrix sparsity:', sparsity)

# Recommend a method based on sparsity
method = recommend_method(sparsity)
print(method)
```
