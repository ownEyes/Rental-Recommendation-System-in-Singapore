```python
import random
import pandas as pd

recommend_default_topn=10
df = pd.read_csv('./dataSource/final_data.csv')


def rating_recommendation(df, recommend_default_topn):
    rating_cols=['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']
    df = df.assign(total_rating=df[rating_cols].sum(axis=1))

    sorted_df = df.sort_values(by='total_rating', ascending=False)
    max_total_rating = sorted_df['total_rating'].max()
    topn_candidates = sorted_df[sorted_df['total_rating'] == max_total_rating]
    

    if len(topn_candidates) <= recommend_default_topn:
        result = topn_candidates
    else:
        sampled_rows = random.sample(topn_candidates.index.tolist(), recommend_default_topn)
        result = df.loc[sampled_rows]
    return result
```

```python
result=rating_recommendation(df, recommend_default_topn)
result
```
